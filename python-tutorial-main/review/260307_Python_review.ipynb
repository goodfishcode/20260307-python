# Python 基本題目練習（10題）

本筆記僅提供題目，不含解答程式碼。

## 第1題：判斷奇偶數

### 題目
- 輸入一個整數，判斷是奇數或偶數。
- 輸出規則：偶數輸出 `Even`，奇數輸出 `Odd`。
- 範例輸入：`7`
- 範例輸出：`Odd`


## 第2題：攝氏轉華氏

### 題目
- 輸入攝氏溫度 `c`（浮點數），轉換為華氏溫度。
- 公式：`F = C * 9 / 5 + 32`。
- 輸出格式：`Fahrenheit =`，並顯示到小數點後一位。


## 第3題：1 到 n 的總和

### 題目
- 輸入正整數 `n`，計算 `1 + 2 + ... + n`。
- 輸出格式：`Sum =` 加上總和。


## 第4題：三數最大值

### 題目
- 輸入三個整數，找出最大值。
- 輸出格式：`Max =` 加上最大值。


## 第5題：正 n 邊形面積

### 題目
- 輸入正數 `n`、`s`，分別代表正 `n` 邊形邊數與邊長。
- 面積公式：`Area = (n * s^2) / (4 * tan(pi / n))`。
- 輸出格式：`Area =`，並顯示到小數點後四位。


## 第6題：List 平均值

### 題目
- 使用者先輸入一個正整數 `n`，再輸入 `n` 個整數存入 `list`。
- 計算平均值並輸出到小數點後二位。
- 輸出格式：`Average = xx.xx`。


## 第7題：List 去除重複並排序

### 題目
- 輸入一行整數（以空白分隔）並存成 `list`。
- 請移除重複元素後由小到大排序。
- 輸出排序後結果（空白分隔）。


## 第8題：字串中母音數量

### 題目
- 輸入一個英文句子（可含空白）。
- 計算母音 `a, e, i, o, u`（大小寫都算）總數。
- 輸出格式：`Vowels =` 加上數量。


## 第9題：Function 計算階乘

### 題目
- 撰寫函式 `factorial(n)`，回傳 `n!`。
- 主程式輸入一個非負整數 `n`，呼叫函式後輸出結果。
- 輸出格式：`n! = result`。


## 第10題：Function 判斷質數

### 題目
- 撰寫函式 `is_prime(n)`，判斷 `n` 是否為質數。
- 主程式輸入一個大於 1 的整數。
- 若是質數輸出 `Prime`，否則輸出 `Not Prime`。
