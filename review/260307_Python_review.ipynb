## 第1題：判斷奇偶數

### 題目
- 輸入一個整數，判斷是奇數或偶數。
- 輸出規則：偶數輸出 `Even`，奇數輸出 `Odd`。
- 範例輸入：`7`
- 範例輸出：`Odd`


```python
num = int(input('輸入一個整數'))
if num % 2 == 0:
    print('Even')
else:
    print('Odd')
```

## 第2題：攝氏轉華氏

### 題目
- 輸入攝氏溫度 `c`（浮點數），轉換為華氏溫度。
- 公式：`F = C * 9 / 5 + 32`。
- 輸出格式：`Fahrenheit =`，並顯示到小數點後一位。


```python
C = float(input('輸入攝氏溫度:'))
F = C * 9 / 5 + 32
print(f'Fahrenheit = {F:.1f}')
```

## 第3題：1 到 n 的總和

### 題目
- 輸入正整數 `n`，計算 `1 + 2 + ... + n`。
- 輸出格式：`Sum =` 加上總和。


```python
n = int(input('輸入正整數:'))
sum = 0
for i in range(n+1):
    sum += i
print(f'Sum = {sum}')
```

## 第4題：三數最大值

### 題目
- 輸入三個整數，找出最大值。
- 輸出格式：`Max =` 加上最大值。


```python
num = int(input('輸入三個整數'))
maximum = 0
for n in range(num):
    if maximum < n:
        maximum = n
print(f'Max = {maximum}')
```

## 第5題：正 n 邊形面積

### 題目
- 輸入正數 `n`、`s`，分別代表正 `n` 邊形邊數與邊長。
- 面積公式：`Area = (n * s^2) / (4 * tan(pi / n))`。
- 輸出格式：`Area =`，並顯示到小數點後四位。


```python
import math
n = int(input())
s = float(input())
area = (n * s**2) / (4 * math.tan(math.pi / n))
print('Area =',f'{area:.4f}')
```

## 第6題：List 平均值

### 題目
- 使用者先輸入一個正整數 `n`，再輸入 `n` 個整數存入 `list`。
- 計算平均值並輸出到小數點後二位。
- 輸出格式：`Average = xx.xx`。


```python
n1 = int(input())
lines= []
for i in range(n1):
    lines.append(int(input()))
total = 0
for ele in lines:
    total += ele
average = total / len(lines)
print(f'Average = {average:.2f}')
```

## 第7題：List 去除重複並排序

### 題目
- 輸入一行整數（以空白分隔）並存成 `list`。
- 請移除重複元素後由小到大排序。
- 輸出排序後結果（空白分隔）。


```python
data = input().split()
data1 = data[:]
sorted(set(data1))
print(f'原始list: {data}')
print(f'新的list: {data1}')
```
