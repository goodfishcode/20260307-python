## 第1題：判斷奇偶數

## 第5題：正 n 邊形面積

### 題目
- 輸入正數 `n`、`s`，分別代表正 `n` 邊形邊數與邊長。
- 面積公式：`Area = (n * s^2) / (4 * tan(pi / n))`。
- 輸出格式：`Area =`，並顯示到小數點後四位。


```python
import math
n = int(input())
s = float(input())
area = (n * s**2) / (4 * math.tan(math.pi / n))
print('Area =',f'{area:.4f}')
```

## 第6題：List 平均值

### 題目
- 使用者先輸入一個正整數 `n`，再輸入 `n` 個整數存入 `list`。
- 計算平均值並輸出到小數點後二位。
- 輸出格式：`Average = xx.xx`。


```python
n1 = int(input())
lines= []
for i in range(n1):
    lines.append(int(input()))
total = 0
for ele in lines:
    total += ele
average = total / len(lines)
print(f'Average = {average:.2f}')
```

## 第9題：Function 計算階乘

### 題目
- 撰寫函式 `factorial(n)`，回傳 `n!`。
- 主程式輸入一個非負整數 `n`，呼叫函式後輸出結果。
- 輸出格式：`n! = result`。


```python
# class factorial:
#     def __init__(self, n):
#         self.n = n
#         for i in range(n+1):
def factorial(n):
    result = 1
    if n > 1:
        for i in range(2, n+1):
            result *= i
    return result

n = int(input())
print(f'{n}! = {factorial(n)}')
```

## 第10題：Function 判斷質數

### 題目
- 撰寫函式 `is_prime(n)`，判斷 `n` 是否為質數。
- 主程式輸入一個大於 1 的整數。
- 若是質數輸出 `Prime`，否則輸出 `Not Prime`。


```python
# 老師解答
def is_prime(n):
    if n <= 1:
        return False
    
    i=2
    while i < n:
        if n % i == 0:
            return False
        i += 1
    return True

x = int(input())
if is_prime(x) == True:
    print(f'{x} = Prime')
else:
    print(f'{x} = Not Prime')
```
